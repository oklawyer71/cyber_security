```python
# Sampling the data, this isn't required, but I just wanted to test it out on a subset of the data to start

subset_benign = df_benign.sample(frac=0.2)
```

```python
# in the above sample / head, we see that there is a mixture of categorical and non-categorical data types (int, float, etc.) so let's check all of the fields

subset_benign.info()
```

```python
# for K-means, we need to have numerical, but for k-modes, we need categorical.  Let's start with K-modes

subset_benign.select_dtypes('object').nunique()
```

```python
# Select the categorical columns
cols = subset_benign.select_dtypes('object').columns
df_cat = subset_benign[cols]
# Show the data
df_cat.head()
```

```python
# let's check what the data looks like in terms of nulls

df_cat.isna().sum()
```

```python
# Setting up the cluster for the k-modes

cluster = setup(df_cat, normalize = True, ignore_features = ['sld'], session_id = 123)
```
